```python
#W3A3
def luythua(n):
    if n <= 0:
        return False
    return (n & (n - 1)) == 0
print(f"16 là lũy thừa của 2? {luythua(16)}")  
print(f"12 là lũy thừa của 2? {luythua(12)}")  
```

```python
#W3A11
a = int(input())
b = int(input())
c = int(input())
if a + b > c and b + c > a and a + c >b :
    if a == b == c:
        print("equilateral triangle")
    elif a == b != c :
        print("isosceles triangle")
    elif a**2 + b**2 == c**2 or b**2 +c**2 == a**2 or a**2 +c**2 == b**2:
        print("right triangle")
    else:
        print("regular triangle")       
else:
    print("not triangle")
```
