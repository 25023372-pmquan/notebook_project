```python
#W5A3
def prime(n):
    if n < 2:
        return False
    for i in range (2, int(n**0.5) + 1):
        if n % i == 0:
            return False     
    return True
n = int(input())
print(prime(n))
```

```python
#W5A4
def perfect_num(n):
    tong = 0
    for i in range (1,int(n/2 + 1)):
        if n % i == 0:
            tong += i
    if tong == n:
        return True
    else:
        return False
n = int(input())
if perfect_num(n):
    print("perfect number")
else:
    print("not perfect number")
```

```python
#W5A6
def factorial(n):
    fac = 1
    for i in range(1,n+1):
        fac *= i
    return fac 
n = int(input("nhap so nguyen duong"))
print(f"giai thua cua {n} là {factorial(n)}")
        
```

```python
#W5A9
def tinh_tong(n):
    tong = 0
    while n > 0:
        tong = tong + n % 10
        n //= 10
    return tong  
n = int(input("nhap so nguyen duong :"))
print(f"tong cac chu so cua {n} la {tinh_tong(n)}")
```

```python
#W5A10
def dangcau(a,b):
    map_ab = {}
    map_ba = {}
    for x,y in zip(a,b):
        if x in map_ab :
            if map_ab[x] != y:
                return False
        else :
            map_ab[x] = y
        if y in map_ba:
            if map_ba[y] != x:
                return False
        else:
            map_ba[y] = x
    return True
a = 'abca'
b = 'zbxz'
print(dangcau(a,b))
```
