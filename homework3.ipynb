```python
#bai 4
height = int(input("height:"))
day = 1 + 2 * (height - 1)
i = 1
while i <= height:
    star = 1 + 2 * (i - 1)
    space = (day - star) // 2
    new_space = star - 2
    if i == 1:
        print (" " * space + "*")
    elif i == height:
        print("*" * day)
    else:
        print(" " * space + "*" + new_space * " " + "*")
    i += 1
```
