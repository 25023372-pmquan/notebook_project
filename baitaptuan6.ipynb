```python
# W6A1
# Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng 
# giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. 
# Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

n = list(map(int,input().split()))
lst = []
for i in n :
    if i not in lst: lst.append(i)
print(lst)
```

```python
# W6A2
# Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự

n = list(map(int,input().split()))
lst = []
tong = 0
for i in n:
    tong += i
    lst.append(tong)
print(lst)
```

```python
# W6A3
# Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple 
# sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)

t = tuple(map(int,input().split()))
k = int(input())
n = t[k:] + t[:k]
print(n)
```

```python
# W6A4
#Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
# dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
# lặp lại, thêm vào list của key đó

n = input().split()
d = {}
for i in n :
    key,value = i.split(":")
    if key not in d: 
        d[key] = [] 
    d[key].append(value)
print(d)
```

```python
# W6A5
# Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)

n = list(map(int,input().split()))
dic = {'positives':0 , 'negatives':0 , 'zeros':0}
for i in n:
    if i > 0: dic['positives'] += 1
    elif i < 0: dic['negatives'] += 1
    else: dic['zeros'] += 1
    
print(dic)
```

```python
# W6A8
# Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary

lst = list(input().split())
dic = {}
for i in lst:
    if i not in dic:
        dic[i] = 1
    else:
        dic[i] += 1
print(dic)
```

```python
# W6A9
# Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một 
# dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ 
# được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới 
# theo thứ tự từ điển tăng dần của key    

def input_dict():
    n = input("nhap dictionary:").split()
    d = {}
    for i in n :
        key,value = i.split(":")
        if key not in d:
            d[key] = value
    return d
d1 = input_dict()
d2 = input_dict()
for key,value in d2.items():
    if key not in d1:
        d1[key] = value
    else :
        d1[key] += value
    
print(d1)


```

```python
# W6A10
# Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên 
# dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là 
# một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh 
# sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được 
# sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

n = list(map(int,input("nhap day so").split()))
k = int(input("nhap so k:"))
lst = []
for a in range (len(n)):
    for b in range (a + 1,len(n)):
        if n[a] + n[b]== k and (a,b) not in lst: 
            lst.append((n[a],n[b]))
print(lst)
            
```

```python
# W6A11
# Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa 
# tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên 
# thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt

t = tuple(map(int,input("nhap day so:").split()))
t1 = ()
t2 = ()
for i in range (len(t)):
    if t[i] % 2 == 0:
        t1 += (t[i],)
    else:
        t2 += (t[i],)
print("even:" ,t1)
print("odd:" ,t2)

```

```python
# W6A12
# [Tìm phần tử xuất hiện nhiều nhất] 
# Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong 
# list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ 
# nhất trong số đó

n = list(map(int,input("nhap day so:").split()))
dic = {}
for i in n:
    if i not in dic:
        dic[i] = 1
    else:
        dic[i] += 1
max_value = max(dic.values())
temp = []
for key in dic:
    if dic[key] == max_value :
        temp.append(key)
if len(temp) == 1:
    print("so xuat hien nhieu nhat la:",*temp)
else:
    print(min(temp))
                
```

```python
# W6A13
# [Đảo ngược key và value của dictionary] 
# Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo 
# một dictionary mới bằng cách đảo ngược key và value của dictionary ban 
# đầu. In ra dictionary mới.

def dictionary():
    d = {}
    i = 1
    while True:
        key = input(f"Nhập key thứ {i}: ")
        value = input(f"Nhập value thứ {i}: ")
        d[key] = value
        i += 1
        n = input("nhap 1 de dung phim khac de tiep tuc")
        if n == '1':
            return d
dic = dictionary()
new_dic = {}
for key, value in dic.items():
    new_dic[value] = key
print(dic)
print(new_dic)
```

```python
# W6A14
# [Giao của hai danh sách] 
# Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các 
# phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các 
# phần tử chung đó, không trùng lặp.
l1 = list(map(int,input().split()))
l2 = list(map(int,input().split()))
lst = []
for a in range (len(l1)):
    for b in range (len(l2)):
        if l1[a] == l2[b] and l1[a] not in lst:
            lst.append(l1[a])
print(lst)

```

```python
# W6A15
# [Lọc dictionary theo giá trị] 
# Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
# nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp 
# key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary 
# mới.

def dictionary():
    d = {}
    i = 1
    while True:
        key = input(f"Nhập string key thứ {i}: ")
        value = input(f"Nhập int value thứ {i}: ")
        d[key] = value
        i += 1
        n = input("nhap 1 de dung phim khac de tiep tuc")
        if n == '1':
            return d
dic = dictionary()
k = int(input("nhap so nguyen"))
new_dic = {}
for key,value in dic.items():
    if int(value) > k:
        new_dic[key] = value
print(new_dic)

```
