```python
# W6A16 
# Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải.

n, m = map(int,input("nhap n hang m cot:").split())
arr = []
for i in range (n):
    arr.append(list(map(int,input(f"nhap {m} so vao hang {i}:").split())))

for row in arr:
    for i in row:
        print(f"{i:4d}", end="")
    print()
```

```python
# W6A18
# Nhâp vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên cùng một dòng, cách nhau dấu cách).
# Tính và in ra tổng các phần tử trên cột k (hàng k)
n,m,k = map(int,input("nhap m, n va k:").split())
arr = []
for i in range (1,n + 1):
    arr.append(list(map(int,input(f"nhap {m} so vao hang {i}:").split())))

tong_cot_k = 0
for i in range (n):
    tong_cot_k += arr[i][k - 1]
print(f"tong cot {k} la {tong_cot_k}")

tong_hang_k = sum(arr[k - 1])
print(f"tong hang {k} la {tong_hang_k}")
```

```python
# W6A20
# Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
# trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

def find_num():
    n = list(map(int,input("nhap day so nguyen:").split()))
    k = int(input("nhap so k:"))
    for i in range(len(n)):
        if n[i] == k:
            return i + 1
    return -1
print(find_num())
```

```python
# W6A21
# Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
# chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

n = list(input().split())
print (n)
dic = {}
for i,word in enumerate(n):
    if word not in dic:
        dic[word] = []
    dic[word].append(i)
for k in dic:
    dic[k] = tuple(dic[k])
print(dic)
```

```python
# W6A22
# Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất trong list

def find_index_max(lst):
    for i in range(len(lst)):
        if lst[i] == max(lst):
            return i
lst = list(map(int,input().split()))
print(lst)
print("vi tri cua so lon nhat:",find_index_max(lst))
    
```

```python
# W6A23 
# VIết hàm xóa đi các số <x có trong mảng

def delete_arr(arr):
    x = int(input("xoa di cac so nho hon :"))
    for i in range(len(arr)):
        arr[i] = [k for k in arr[i] if k >= x]
    return arr
    
n, m = map(int,input("nhap n hang m cot:").split())
arr = []
for i in range (n):
    arr.append(list(map(int,input(f"nhap {m} so vao hang {i + 1}:").split())))
print(delete_arr(arr)) 
        
```

```python
# W6A24
# Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí 
# trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

def nnn(n):
    lst = []
    for i in range(len(n)):
        if n[i] == 0:
            lst.append(0)
    n = [k for k in n if k != 0]
    return lst + n

n = list(map(int,input().split()))
print(*nnn(n))
```

```python
# W6A25
# Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong mảng
def arr_input():
    arr = []
    while True:
        arr.append(list(map(int,input("nhap vao mang").split())))
        k = input("nhap q de dung phim khac de tiep tuc")
        if k == 'q':
            return arr
n = list(map(int,input("nhap day so").split()))
arr = arr_input()
for i in range(len(arr)):
    for j in range (len(arr[i])):
        if arr[i][j] in n:
            n = [x for x in n if x != arr[i][j]]
print("nhung gia tri khong co trong mang:",*n)
            
```

```python
# W6A26
# Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

lst = list(map(int,input().split()))
result = 0
for i in range (1,len(lst)):
    if lst[i] > lst[i - 1] and lst[i] >lst[i + 1]:
        result += lst[i]
print(result)
```

```python
# W6A27
# Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1

lst = list(map(int,input().split()))
x = int(input())
l = []
for i in range(len(lst)):
    if lst[i] < x:
        l.append(lst[i])
if len(l) == 0:
    print("-1")
else:      
    print(*l)
```

```python
# W6A28
# Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 
# thực sự, hay vô danh (không tăng, không giảm)

lst = list(map(int,input().split()))
temp1 = []
temp2 = []
for i in range(1,len(lst)):
    if lst[i] > lst[i - 1]:
        temp1.append(1)
    if lst[i] < lst[i - 1]:
        temp2.append(1)
if len(temp1) == len(lst) - 1:
    print("la chuoi thuc su tang")
elif len(temp2) == len(lst) - 1:
    print("la chuoi thuc su giam")
else:
    print("la chuoi vo danh")
    
        
        
    
```
