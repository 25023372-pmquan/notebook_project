```python
#bai 2
try:
    n=0
    for _ in range(10000):
        n = int(input("nhap so nguyen duong"))
        if n > 0:
            break 
        else:
            print("ban phai nhap so nguyen duong de dung!")
if n < 2:
    print("Không phải số nguyên tố")
else:
    la_nguyen_to = True
    for i in range(2, n):
        if n % i == 0:
            la_nguyen_to = False
            break       
    if la_nguyen_to:
        print("Là số nguyên tố")
    else:
        print("Không phải số nguyên tố")
except ValueError:
    print("nhap sai")
```

```python
#bai 4
n = int(input("Nhập n: "))
if n < 0:
    n = -n
if n == 0:
    dem = 1
else:
    while n > 0:
        n //= 10
        dem += 1
print("Số chữ số:", dem)
```

```python
#bai 11
n = int(input("Nhập n: "))
dem = 0
for i in range(2, n + 1, 2):
    if n % i == 0:
        dem += 1
print("Số ước chẵn:", dem)
```
