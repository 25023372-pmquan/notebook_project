```python
#bai 7 
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input("Nhập n: "))
temp = n
while True:
    if(is_prime(temp)) and n%temp==0:
        break
    temp -= 1   
print(temp)
```

```python
#bai 8
def reverse(n):
    rev = 0
    while n > 0:
        rev = rev*10 + n%10
        n //= 10
    return rev
n = int(input())
step = 0
while True:
    if n == reverse(n):
        break
    n = n + reverse(n)
    step += 1
print("step:",step)
print("palindrome:",n)
    
```

```python
#bai 9
def is_square(n):
    if n < 0:
        return False
    x = int(n**0.5)
    return x*x == n
def digits_distinct(n):
    s = str(n)
    return len(set(s)) == len(s)
found = False
n = int(input())
for i in range (0,n+1):
    if is_square(i) and digits_distinct(i):
        print(i,end=" ")
        found = True
if not found:
    print("no number")
```
